# KPI Analysis

To evaluate customer retention and relationship strength, several Key Performance Indicators (KPIs) were developed and analyzed. These KPIs helped in understanding how customer engagement, product utilization, and financial commitment influence customer loyalty.



## 1. Engagement Retention Ratio

This KPI compares the churn rates of active and inactive customers. The objective is to understand whether customer engagement influences customer retention.

```python
active_retention = 85.730926
inactive_retention = 73.149103

ERR = active_retention / inactive_retention

print("Engagement Retention Ratio:", round(ERR,2))
```

```python
retention_rate = (1-df['Exited'].mean())*100

print("Retention Rate:", round(retention_rate,2), "%")
```

2. Product Depth Index

The Product Depth Index was assessed using the number of banking products utilized by customers. Customers using two products showed the highest retention rate, suggesting that greater product adoption strengthens the customer-bank relationship.

```python
product_churn = pd.crosstab(
    df['NumOfProducts'],
    df['Exited'],
    normalize='index'
)*100

print(product_churn)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

sns.barplot(
    x='NumOfProducts',
    y='Exited',
    data=df
)

plt.title("Product Count vs Churn Rate")
plt.show()
```

3. High-Balance Disengagement Rate

This KPI was used to identify customers who maintain high account balances but remain inactive. The findings showed that a substantial proportion of premium customers were disengaged, highlighting a potential risk of silent churn.

```python
high_balance = df['Balance'].quantile(0.75)

premium_customers = df[df['Balance'] >= high_balance]

at_risk_premium = premium_customers[
    premium_customers['IsActiveMember'] == 0
]

print("Premium Customers:", len(premium_customers))
print("At-Risk Premium Customers:", len(at_risk_premium))
```

```python
risk_rate = len(at_risk_premium) / len(premium_customers) * 100

print("High-Balance Disengagement Rate:", round(risk_rate,2), "%")
```

4. Credit Card Stickiness Score

The impact of credit card ownership on retention was evaluated. The results indicated only a marginal difference in churn rates between customers with and without credit cards, suggesting that card ownership alone is not a strong retention driver.

```python
card_churn = pd.crosstab(
    df['HasCrCard'],
    df['Exited'],
    normalize='index'
)*100

card_churn
```

5. Relationship Strength Index (RSI)

A Relationship Strength Index was created by combining customer activity status and product utilization levels. Customers with higher RSI scores demonstrated lower churn rates, indicating stronger relationships with the bank and higher retention potential.

```python
df['RSI'] = (
    0.4 * df['IsActiveMember']
    + 0.4 * (df['NumOfProducts'] / 4)
    + 0.2 * df['HasCrCard']
)

df['RSI'].describe()
```

Overall, the KPI analysis confirms that customer engagement and product utilization are stronger indicators of retention than financial value alone. These findings can support the development of targeted retention strategies and customer relationship management initiatives.

Customer Segmentation

```python
df['CustomerSegment'] = 'Other'

# Sticky Customers
df.loc[
    (df['IsActiveMember']==1) &
    (df['NumOfProducts']>=2),
    'CustomerSegment'
] = 'Sticky Customer'

# At-Risk Premium
df.loc[
    (df['IsActiveMember']==0) &
    (df['Balance']>=df['Balance'].quantile(0.75)),
    'CustomerSegment'
] = 'At-Risk Premium'

# Cross-Sell Opportunity
df.loc[
    (df['IsActiveMember']==1) &
    (df['NumOfProducts']==1),
    'CustomerSegment'
] = 'Cross-Sell Opportunity'

# High Churn Risk
df.loc[
    (df['IsActiveMember']==0) &
    (df['NumOfProducts']==1),
    'CustomerSegment'
] = 'High Churn Risk'

df['CustomerSegment'].value_counts()
```

Executive Summary

This project analyzes customer engagement and product utilization to identify key factors influencing customer retention in the banking sector. Using a dataset of 10,000 customers from the European banking market, the study examines customer activity levels, product adoption, account balances, demographic factors, and churn behavior.

The analysis revealed that customer engagement is a major driver of retention. Active customers exhibited a significantly lower churn rate (14.27%) compared to inactive customers (26.85%). Product utilization also played a critical role, with customers holding two banking products showing the highest retention rate (92.42%).

Geographical analysis indicated that Germany experienced the highest churn rate (32.44%), suggesting the need for region-specific retention strategies. Furthermore, customers who churned maintained higher average balances than retained customers, demonstrating that financial value alone does not guarantee loyalty.

The study identified several customer segments, including Sticky Customers, Cross-Sell Opportunities, High Churn Risk Customers, and At-Risk Premium Customers. These segments provide actionable insights for designing targeted retention campaigns, improving product bundling strategies, and strengthening customer relationships.

Overall, the findings suggest that customer engagement and relationship depth are stronger predictors of retention than traditional financial indicators, supporting the development of engagement-driven retention strategies.

Key Recommendations
1. Increase engagement initiatives for inactive customers.
2. Promote multi-product adoption through cross-selling campaigns.
3. Develop specialized retention programs for high-balance inactive customers.
4. Prioritize customer retention efforts in Germany.
5. Use the Relationship Strength Index (RSI) as an early-warning indicator of churn.
6. Design personalized banking experiences for customers above 40 years of age.











Conclusion

The analysis demonstrates that customer engagement and relationship depth are stronger predictors of retention than financial value alone.

Active customers and customers using multiple banking products exhibit significantly lower churn rates. Conversely, inactive customers, single-product customers, and high-balance disengaged customers represent the greatest retention risk.

The findings support the implementation of engagement-focused retention strategies, targeted cross-selling initiatives, and personalized relationship management programs to improve customer loyalty and reduce churn.
